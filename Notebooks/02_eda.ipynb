```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\adaml\\Documents\\Stage\\Pricing engine\\Data\\processed\\data.csv')  
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation with checkout_price
corr = data.corr(numeric_only=True)[['checkout_price','num_orders']].sort_values(by='checkout_price', ascending=False)

plt.figure(figsize=(4, 10))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    center=0,
    fmt='.2f'
)

plt.title('Correlation with checkout_price')
plt.show()
```
